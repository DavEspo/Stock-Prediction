```python
import pandas as pd
stock_data = pd.read_csv('stocks.csv')
stock_desc_feat = ['Ticker', 'Date', 'Open', 'High', 'Low', 'Close', 'Volume']
# stock_data_features = stock_data.columns
stock_data.head()
```

```python
historical_data = pd.read_csv('data.csv')
hist_desc_feat = ['Company', 'Date', 'Volume', 'Open', 'High', 'Low']
historical_data.head()
```

```python
tesla_stock = pd.read_csv('tsla_2014_2023.csv')
tesla_desc_feat = ['date', 'open', 'high', 'low', 'volume', 'rsi_7', 'rsi_14', 'cci_7', 'cci_14', 'sma_50', 'ema_50', 'sma_100', 'ema_100', 'macd', 'bollinger', 'TrueRange', 'atr_7', 'atr_14', 'next_day_close']
tesla_stock.head()
```

```python
print(stock_data['Ticker'].unique())
print(historical_data['Company'].unique(), "\n")
print(stock_data.isnull().sum(), "\n")
print(historical_data.isnull().sum(), "\n")
print(tesla_stock.isnull().sum())
```

```python
# tesla_stock[''].unique()
```

```python
historical_data['Close/Last'] = historical_data['Close/Last'].str.replace('$', '', regex=False)
historical_data['Open'] = historical_data['Open'].str.replace('$', '', regex=False)
historical_data['High'] = historical_data['High'].str.replace('$', '', regex=False)
historical_data['Low'] = historical_data['Low'].str.replace('$', '', regex=False)
historical_data['Close/Last'] = pd.to_numeric(historical_data['Close/Last'])
historical_data['Open'] = pd.to_numeric(historical_data['Open'])
historical_data['High'] = pd.to_numeric(historical_data['High'])
historical_data['Low'] = pd.to_numeric(historical_data['Low'])
print(historical_data.dtypes)
historical_data
```

```python
# Splitting the data into training and testing (33% testing)
from sklearn.model_selection import train_test_split

X_train_stock, X_test_stock, Y_train_stock, Y_test_stock = train_test_split(stock_data[stock_desc_feat], stock_data['Adj Close'], test_size=.33, random_state=0)
X_train_hist, X_test_hist, Y_train_hist, Y_test_hist = train_test_split(historical_data[hist_desc_feat], historical_data['Close/Last'], test_size=.33, random_state=0)
X_train_tesla, X_test_tesla, Y_train_tesla, Y_test_tesla = train_test_split(tesla_stock[tesla_desc_feat], tesla_stock['close'], test_size=.33, random_state=0)
```

```python
print(stock_data.dtypes, "\n")
print(historical_data.dtypes, "\n")
print(tesla_stock.dtypes)
```

```python
# Categorical features made into continuous in the training and testing sets through encoding
from sklearn.preprocessing import OrdinalEncoder

cat_feat_stock = ['Ticker', 'Date']
cat_feat_hist = ['Company', 'Date']
cat_feat_tesla = ['date']

enc=OrdinalEncoder()

# 
enc.fit(X_train_stock[cat_feat_stock])
X_train_stock[cat_feat_stock]=enc.transform(X_train_stock[cat_feat_stock])

enc.fit(X_test_stock[cat_feat_stock])
X_test_stock[cat_feat_stock]=enc.transform(X_test_stock[cat_feat_stock])

# 
enc.fit(X_train_hist[cat_feat_hist])
X_train_hist[cat_feat_hist]=enc.transform(X_train_hist[cat_feat_hist])

enc.fit(X_test_hist[cat_feat_hist])
X_test_hist[cat_feat_hist]=enc.transform(X_test_hist[cat_feat_hist])

# 
enc.fit(X_train_tesla[cat_feat_tesla])
X_train_tesla[cat_feat_tesla]=enc.transform(X_train_tesla[cat_feat_tesla])

enc.fit(X_test_tesla[cat_feat_tesla])
X_test_tesla[cat_feat_tesla]=enc.transform(X_test_tesla[cat_feat_tesla])
```

```python
# Normalization of training and testing sets
from sklearn.preprocessing import MinMaxScaler

# Stock Data
NormalizedDF=MinMaxScaler().fit(X_train_stock)
X_train_stock=NormalizedDF.transform(X_train_stock)
X_train_stock=X_train_stock.round(2)

NormalizedDF=MinMaxScaler().fit(X_test_stock)
X_test_stock=NormalizedDF.transform(X_test_stock)
X_test_stock=X_test_stock.round(2)

# Historical Data
NormalizedDF=MinMaxScaler().fit(X_train_hist)
X_train_hist=NormalizedDF.transform(X_train_hist)
X_train_hist=X_train_hist.round(2)

NormalizedDF=MinMaxScaler().fit(X_test_hist)
X_test_hist=NormalizedDF.transform(X_test_hist)
X_test_hist=X_test_hist.round(2)

# Tesla Dataset
NormalizedDF=MinMaxScaler().fit(X_train_tesla)
X_train_tesla=NormalizedDF.transform(X_train_tesla)
X_train_tesla=X_train_tesla.round(2)

NormalizedDF=MinMaxScaler().fit(X_test_tesla)
X_test_tesla=NormalizedDF.transform(X_test_tesla)
X_test_tesla=X_test_tesla.round(2)
```

```python
# Artificial Neural Networks

from sklearn.metrics import accuracy_score
```

```python
# ARIMA
```

```python
# Gradient Boosting
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score

# SEED = 23

gbr = GradientBoostingRegressor(n_estimators = 300, learning_rate = 0.05, random_state = 100, max_features = 5)

gbr.fit(X_train_stock, Y_train_stock)
pred_Y_stock = gbr.predict(X_test_stock)
mse_stock = mean_squared_error(Y_test_stock, pred_Y_stock)
r2_stock = r2_score(Y_test_stock, pred_Y_stock)
print(f"Gradient Boosting Regressor MSE for stock is: {mse_stock:.2f}")
print(f"Gradient Boosting Regressor R2 for stock is: {r2_stock:.2f}")
# print(f"Gradient Boosting Classifier accuracy is : {acc_stock:.2f}")

gbr.fit(X_train_hist, Y_train_hist)
pred_Y_hist = gbr.predict(X_test_hist)
mse_hist = mean_squared_error(Y_test_hist, pred_Y_hist)
r2_hist = r2_score(Y_test_stock, pred_Y_stock)
print(f"Gradient Boosting Regressor MSE for hist is: {mse_hist:.2f}")
print(f"Gradient Boosting Regressor R2 for hist is: {r2_hist:.2f}")
# print(f"Gradient Boosting Classifier accuracy is : {acc_hist:.2f}")

gbr.fit(X_train_tesla, Y_train_tesla)
pred_Y_tesla = gbr.predict(X_test_tesla)
mse_tesla = mean_squared_error(Y_test_tesla, pred_Y_tesla)
r2_tesla = r2_score(Y_test_stock, pred_Y_stock)
print(f"Gradient Boosting Regressor MSE for tesla is: {mse_tesla:.2f}")
print(f"Gradient Boosting Regressor R2 for tesla is: {r2_tesla:.2f}")
# print(f"Gradient Boosting Classifier accuracy is : {acc_tesla:.2f}")
```
