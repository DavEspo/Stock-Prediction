```python
# Make the columns that are supposed to be float to float
historical_data['Close/Last'] = historical_data['Close/Last'].str.replace('$', '', regex=False)
historical_data['Open'] = historical_data['Open'].str.replace('$', '', regex=False)
historical_data['High'] = historical_data['High'].str.replace('$', '', regex=False)
historical_data['Low'] = historical_data['Low'].str.replace('$', '', regex=False)
historical_data['Close/Last'] = pd.to_numeric(historical_data['Close/Last'])
historical_data['Open'] = pd.to_numeric(historical_data['Open'])
historical_data['High'] = pd.to_numeric(historical_data['High'])
historical_data['Low'] = pd.to_numeric(historical_data['Low'])
print(historical_data.dtypes)
historical_data
```

```python
# ARIMA
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# For ARIMA we need series only (not features)
# Assuming Y_train_stock, Y_test_stock are pandas Series and properly ordered

def fit_arima_and_evaluate(train, test, label):
    model = ARIMA(train, order = (5, 1, 0)) # (p, d, q) - can be tuned
    model_fit = model.fit()
    forecast = model_fit.forecast(steps = len(test))
    mse = mean_squared_error(test, forecast)
    r2 = r2_score(test, forecast)
    print(f"ARIMA MSE for {label} is: {mse:.2f}")
    print(f"ARIMA R2 for {label} is: {r2:.2f}")
    return forecast

forecast_stock = fit_arima_and_evaluate(Y_train_stock, Y_test_stock, "stock")
forecast_hist = fit_arima_and_evaluate(Y_train_hist, Y_test_hist, "hist")
forecast_tesla = fit_arima_and_evaluate(Y_train_tesla, Y_test_tesla, "tesla")
```
